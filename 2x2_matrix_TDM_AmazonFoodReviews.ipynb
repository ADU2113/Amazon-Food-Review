```python
from nltk.corpus import stopwords
l1 = list(stopwords.words('english'))
```

```python
import string
string.punctuation
```

```python
def text_process(mess):
    global nopunc
    nopunc = [char for char in mess if char not in string.punctuation ]
    nopunc = "".join(nopunc)

    return[word for word in nopunc.split()if word not in l1]
```

```python
#ta['A'].apply(text_process)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow_transformer = CountVectorizer(analyzer=text_process).fit(df['Text'])
```

```python
bow_transformer.vocabulary_
```

```python
len(bow_transformer.vocabulary_)
```

```python
TDM = bow_transformer.transform(df['Text'])
```

```python
TDM.shape
```

```python
type(TDM)
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x, train_y, test_y = train_test_split(TDM,df.Score)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(train_x , train_y)
pred_test = logreg.predict(test_x)
pred_test
from sklearn.metrics import confusion_matrix
tab1 = confusion_matrix(test_y,  pred_test)
tab1
```

```python
from sklearn.metrics import classification_report
print(classification_report(test_y,pred_test))
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(train_x, train_y)
pred_nb = nb.predict(test_x)
from sklearn.metrics import confusion_matrix
tab_nb = confusion_matrix(test_y, pred_nb)
tab_nb
```

```python
from sklearn.metrics import classification_report
print(classification_report(test_y,pred_nb))
```
